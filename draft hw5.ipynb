```python
#3 Loops
numbers = [1, 2, 3, 4, 5]
def sumWithLoop(numbers):
    total = 0
    for i in numbers:
        total += i
    return total
sumWithLoop(numbers)
#Source: https://stackoverflow.com/questions/13909052/how-do-i-add-together-integers-in-a-list-sum-a-list-of-numbers-in-python
```

```python
#4.1 Lists
null = []
list = ["a", "b", "c"]
def duplicatelist(list):
    for i in list:
        for j in range(0,2):
            null.append(i)
    return null
duplicatelist(list) 
```
