```python
# HW8
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
#1. Introduction to Matplotlib
def make_sine_wave(x, A, w):
    y = A * np.sin(w*x)
    return y
```

```python
#part 2
x = np.linspace(0, 2*np.pi, num = 1000)
```

```python
# A_array = np.linspace(0.5, 2.5, num = 5)
A_array = [0.5, 1.0, 1.5, 2.0, 2.5]
A_array
```

```python
# w_array = np.arange(1, 6)
w_array = [1, 2, 3, 4, 5]
w_array
```

```python
#part 4
figure = plt.figure(figsize = (20, 20))
```

```python
# def sine()
for A, w in zip(A_array, w_array):
    y = A * np.sin(x*w)
    plt.plot(x,y, label = f"A = {A} , w = {w}")

plt.title("Sine Waves")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid()
plt.show()
```

```python
#3 Seaborn!
import seaborn as sns
penguins = sns.load_dataset("penguins")
penguins
```

```python
fig, ax = plt.subplots(1, 2, figsize = (10, 5))
sns.scatterplot(data = penguins, x = "species", y = "body_mass_g", hue = "species", ax = ax[0])
ax[0].set_title("Penguin Species and Body Mass")
ax[0].set_xlabel("Species")
ax[0].set_ylabel("Body Mass (grams)")

sns.histplot(data = penguins, x = "body_mass_g", hue = "species", multiple = "stack")
ax[1].set_title("Penguin Body Mass")
ax[1].set_xlabel("Body Mass (grams)")
ax[1].set_ylabel("Number of Penguins")
```
