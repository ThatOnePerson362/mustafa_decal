```python
def is_prime(i):
    if i == 2:
        return True
    if i == 3:
        return True
    if i == 5:
        return True
    if i == 7:
        return True
    elif i % 2 == 0:
        return False
    elif i % 3 == 0:
        return False
    elif i % 5 == 0:
        return False
    elif i%7 == 0:
        return False
    else:
        return True
is_prime(5)
        
    
```

```python
for i in range(18):
    is_prime(i)
    print(is_prime(i))
```

```python
arr = np.array([[2, 3, 5], [4, 6, 8], [11, 13, 17], [7, 10, 13]])
def containsPrimes(arr):
    keep_row = []
    for row in arr:
        test = False
        for i in row:
            if is_prime(i):
                test = True
        if test:
            keep_row.append(row)
    keep_row = np.array(keep_row)
    return keep_row
print(containsPrimes(arr))
```

```python
#2 Let's Play Checkers
def checkerboard():
    board = np.zeros((8,8))
    return board
checkerboard()
```

```python
#2.2 
def checkerboard():
    board = np.zeros((8,8))
    board[0:9:2, 0:9:2] = 1
    return board
checkerboard()
```

```python
#2.3
def checkerboard():
    board = np.zeros((8,8))
    board[0:9:2, 0:9:2] = 1
    board2 = np.zeros((8,8))
    board2[1:9:2, 1:9:2] = 1
    return board + board2
checkerboard()
```

```python
#2.4
def reverse_checkerboard():
    board = np.zeros((8,8))
    board[1:9:2, 0:9:2] = 1
    board2 = np.zeros((8,8))
    board2[0:9:2, 1:9:2] = 1
    return board + board2
reverse_checkerboard()
```

```python
#3 The Expanding Universe
universe = np.array(["galaxy", "clusters"])
def expanding(universe):
    universe2  = np.char.join(" ", universe)
    return universe2
expanding(universe)
#Source: https://www.w3resource.com/python-exercises/numpy/python-numpy-string-exercise-5.php
```
